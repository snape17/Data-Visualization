In this question, you need to visualize the content of iron, vitamin b1 & b2, niacin for the following commodities:
- Rice
- Pound Cake
- Milk
- Liver (Beef)
- Pork Loin Roast
- Apples

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns



datatable = pd.read_csv('diet.csv')


columns = ['commodity', 'iron_mg' , 'vitamin_b1_mg' , 'vitamin_b2_mg' , 'niacin_mg']
rows= ['Rice' , 'Pound Cake' , 'Milk' ,  'Liver (Beef)' , 'Pork Loin Roast' , 'Apples' , 'Potatoes']
sub_table = datatable.loc[ :, columns]
temp = sub_table.set_index('commodity')
df2 = temp.loc[rows ,:]
df2 = df2.transpose()
print(df2)

fig, ax = plt.subplots(figsize=(15, 8))
sns.heatmap(df2, annot = True, cmap="coolwarm", vmin= 0, vmax=350,
           linewidth= 1.0 , cbar_kws={"shrink": .8}, fmt = 'g' , annot_kws={"size": 18})

plt.xticks(fontsize= 20 , rotation=45)
plt.xlabel('Commodity' , fontsize=18)
plt.ylabel('Nutrient Content' , fontsize=18)

plt.show()
```
